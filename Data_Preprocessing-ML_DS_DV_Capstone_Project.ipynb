## Methods to replace missing values

```python
#1) Replace the missing values with central tendancy (Mean, Median, Mode)
#2) Delete the entire row.
#3) Respective to the problem, have to replace the missing values with hard coded(standard) values. For our data set we need to replace it with 0 whic will not change the meaning or value.
#4) Consider this as a semi supervised, create a model, so that we can predict the missing values.

#For categorical data replacement
#1) Replace with mode
#2) Semi supervised - classification

#Incase of multiple columns NULL values replacement we need use "IMPUTATORS"
```

```python
# In our dataset the NULL values are exists in the Time Taken column and also in the time taken column already 0 also exists. 
# So decided to replace the NULL values with 0 which will not impact the model creation processing. 
dataset["Time_taken"].fillna(0,inplace=True) #inplace=True means the changes directly applied into the dataset, not assigning to any variable.
```
